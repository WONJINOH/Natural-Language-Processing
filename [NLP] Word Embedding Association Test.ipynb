# Measuring Genre Bias Using Word Embedding Association Test (WEAT)


### Concept Review

#### Word Embedding Association Test (WEAT)
This test uses cosine similarity to measure the similarity between two vectors. Cosine similarity measures the cosine of the angle between two vectors. Given two vectors i and j, cosine similarity cos(θ) can be computed using the dot product and magnitude.


$$\frac{mean_x\in X^s(x,A,B)−mean_y\in Y^s(y,A,B)}{std\_dev_{w\in X \cup Y}^{s(w,A,B)}}$$

- X, Y: target word sets (words we want to measure bias for)
- A, B: attribute word sets (concept axes representing bias direction)
- s(w, A, B): The difference in mean cosine similarity between word w and each attribute set

Now let's read the synopsis.txt file and keep only words tagged as nouns, saving them in a variable called `tokenized`. We'll use the konlpy package.


```python
# 약 15분정도 걸립니다.
from konlpy.tag import Okt
okt = Okt()
tokenized = []
with open(os.getenv('HOME')+'/aiffel/weat/synopsis.txt', 'r') as file:
    while True:
        line = file.readline()
        if not line: break
        words = okt.pos(line, stem=True, norm=True)
        res = []
        for w in words:
            if w[1] in ["Noun"]:      # "Adjective", "Verb" 등을 포함할 수도 있습니다.
                res.append(w[0])    # 명사일 때만 tokenized 에 저장하게 됩니다. 
        tokenized.append(res)
```

### STEP 3. Create Target and Attribute Word Sets
In the previous step, we used TF-IDF to create word sets. This method can select representative words to some extent. However, there was a problem of duplicate words appearing despite selecting words with high TF-IDF scores. Poorly selected concept-axis words negatively affect WEAT calculation results.

```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from konlpy.tag import Okt

art_txt = 'synopsis_art.txt'
gen_txt = 'synopsis_gen.txt'

def read_token(file_name):
    okt = Okt()
    result = []
    with open(os.getenv('HOME')+'/aiffel/weat/'+file_name, 'r') as fread: 
        print(file_name, '파일을 읽고 있습니다.')
        while True:
            line = fread.readline() 
            if not line: break 
            tokenlist = okt.pos(line, stem=True, norm=True) 
            #pos(text ) :  텍스트에서 품사 정보를 부착하여 반환한다
            for word in tokenlist:
                if word[1] in ["Noun"]:#, "Adjective", "Verb"]:
                    result.append((word[0])) 
    return ' '.join(result)
```

### STEP 4. Calculate WEAT Score and Visualize

Let's measure bias by movie category and genre, calculate WEAT scores, and visualize them as a heatmap. It would be ideal if the visualization clearly highlights which movie genre attribute combinations show prominent bias.

### Calculate WEAT Score Using the Embedding Model and Word Sets


```python
def cos_sim(i, j):
    return dot(i, j.T)/(norm(i)*norm(j))

def s(w, A, B):
    c_a = cos_sim(w, A)
    c_b = cos_sim(w, B)
    mean_A = np.mean(c_a, axis=-1)
    mean_B = np.mean(c_b, axis=-1)
    return mean_A - mean_B #, c_a, c_b

def weat_score(X,Y,A,B):
    s_X = s(X,A,B)
    s_Y = s(Y,A,B)
    
    mean_X = np.mean(s_X)
    mean_Y = np.mean(s_Y)
    
    std_dev = np.std(np.concatenate([s_X, s_Y], axis = 0))
    
    return (mean_X - mean_Y)/std_dev
```

```python
matrix = [[0 for _ in range(len(genre_name))] for _ in range(len(genre_name))]
```

```python
X = np.array([model.wv[word] for word in target_art])
Y = np.array([model.wv[word] for word in target_gen])

for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        A = np.array([model.wv[word] for word in attributes[i]])
        B = np.array([model.wv[word] for word in attributes[j]])
        matrix[i][j] = weat_score(X, Y, A, B)
```

```python
for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        if matrix[i][j] > 0.8 or matrix[i][j] < -0.8:
            print(genre_name[i], genre_name[j],matrix[i][j])
```
